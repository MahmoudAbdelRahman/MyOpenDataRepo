```python
true_x_0 = 2.0    #m
true_y_0 = 0.0    #m

initial_vx = 5.0 #m/s
initial_vy = 50.0 #m/s

ax = 0.0          #m/s^2
ay = -9.8         #m/s^2

delta_t = 0.1     #sec
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import random
%matplotlib inline
```

```python
xs = [true_x_0]
ys = [true_y_0]

x_measured = [true_x_0]
y_measured = [true_y_0] 

x = 0.0
y = 0.0

v_xs=[5.0]
v_ys=[50.0]

error_x = [0.]
error_y = [0.]

fig, ax = plt.subplots(1, 1, figsize=(5, 10))
for i in range(110):
    if  y>=0 :
        plt.cla()
        t = delta_t*float(i)
        x = initial_vx * t + 2.0
        y = (0.5*-9.8*t*t) + initial_vy * t + 0.0
        vy = (-9.8*t) + initial_vy
        
        xs.append(x)
        ys.append(y)
        
        v_xs.append(5.0)
        v_ys.append(vy)
        
        erx = random.normalvariate(0, 2)
        ery = random.normalvariate(0, 2)
        
        x_measured.append(x + erx)
        y_measured.append(y + ery)
        
        error_x.append(erx)
        error_y.append(ery)

        ax.set_xlim((0, 80))
        ax.set_ylim((0, 160))
        
        ax.plot(xs, ys, color="blue", label="groundTruth")
        ax.plot(x_measured, y_measured, color="black", label="measured", alpha=0.8)
        
        p0 = mpatches.Circle((x, y), radius=1.0, color="red", label="moving object")
        
        ax.add_patch(p0)
        ax.legend()
        plt.savefig("im/something_" + str(i).zfill(3)+".jpg")
plt.close()

df = pd.DataFrame({"X_groundTruth":xs,
                   "Y_groundTruth":ys,
                   "X_measured":x_measured,
                   "Y_measured":y_measured,
                   "Vx":v_xs,
                   "Vy":v_ys,
                  "X_error":error_x,
                  "Y_error":error_y})
df.to_csv("df.csv", index=False)
!convert -loop 20 im/*.jpg im/01_Projectile.gif
!rm -r im/*.jpg
```
